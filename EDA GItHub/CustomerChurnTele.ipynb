Converted values of senior citizen from 0/1 to yes/no for better understanding

```python
def conv(value):
    if value == 1:
        return 'Yes'
    else:
        return 'No'
prime_df['SeniorCitizen'] = prime_df['SeniorCitizen'].apply(conv)
prime_df.head()
```

```python
ax = sns.countplot(x=prime_df['Churn'],data = prime_df)
ax.bar_label(ax.containers[0])
plt.title("Count Of Customers By Churn")
plt.show()
```

```python
plt.figure(figsize=[3,4])
gb = prime_df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'],labels = gb.index,autopct = "%1.2f%%")
plt.title("Percentage Of Customers By Churn")
plt.show()
```

From the pie chart we can conclude that 26.54% of our customers have churned out , now lets explore the reason behind it

```python
plt.figure(figsize=[4,5])
ax =  sns.countplot(x= prime_df['gender'],data = prime_df,hue='Churn')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.ylabel('Total Customers')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(6,6))
total_counts = prime_df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()
total_counts.plot(kind='bar',stacked=True,ax=ax,color= ['#1f77b4','#ff7f0e'])

for p in ax.patches:
    width,height = p.get_width(),p.get_height()
    x,y = p.get_xy()
    ax.text(x+width/2,y+height/2,f'{height:.1f}%',ha='center',va='center')

plt.legend(title='Churn',loc = 'upper right')
plt.show()
```

Comparatively a larger percentage of people from the customers from the senior citizen category have churned out

```python
fig,ax=plt.subplots(figsize=(4,4))
contract_churn = prime_df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()
contract_churn.plot(kind='bar',stacked=True,ax=ax,color= ['#1f77b4','#ff7f0e'])

for p in ax.patches:
    width,height = p.get_width(),p.get_height()
    x,y = p.get_xy()
    ax.text(x+width/2,y+height/2,f'{height:.1%}',ha='center',va='center')
plt.legend()
plt.show()
```

A large percentage of people who were on a month_to_month contract have churned out.
We can create stratergies that cater a bit more to people that are on a monthly subscription or we can convince them to switch on to an annual subscription to gain access to more lucrative features.

```python
services = prime_df.columns[6:14]
plt.figure(figsize=(16, 12))

for i, col in enumerate(services):
    ax = plt.subplot(2, 4, i + 1)
    counts_df = prime_df.groupby(col)['Churn'].value_counts().unstack()

    counts_df.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy()
        
        if height > 0:
            ax.text(x + width / 2,
                    y + height / 2,
                    f'{height:.0f}',
                    ha='center',
                    va='center')

    plt.title(f'Churn by {col}')
    plt.xticks(rotation = 0)

plt.tight_layout()
plt.show()
```

This analysis of customer churn across various service subscriptions highlights critical patterns in customer loyalty. A primary finding is that customers with Fiber optic internet service show a significantly higher propensity to churn compared to those with DSL. Conversely, supplementary services like Online Security, Online Backup, Device Protection, and Tech Support act as strong retention drivers. Customers who do not subscribe to these value-added services are considerably more likely to discontinue their patronage. This suggests that while premium internet offerings attract users, their retention is heavily influenced by their integration into a wider ecosystem of protective services.

```python
pi = prime_df.columns[3:5]
plt.figure(figsize=(15, 5))

for i, col in enumerate(pi):
    plt.subplot(1, 2, i + 1)
    counts = prime_df[col].value_counts()
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

This demographic analysis highlights the relationship and dependency status of the customer base. The data shows an almost even split in partnership status, with 51.7% of customers being single and 48.3% having a partner. In contrast, there is a clear distinction regarding dependents, where a substantial majority of customers (70%) report having no dependents, while the remaining 30% do. This profile suggests a clientele that is balanced in relationships but predominantly consists of individuals or couples without dependents.

```python
fig,ax = plt.subplots(figsize=(5,6))
payment_type = prime_df.groupby('PaymentMethod')['Churn'].value_counts(normalize=True).unstack()
payment_type.plot(kind='bar',stacked=True,ax=ax)
for p in ax.patches:
    width,height = p.get_width(),p.get_height()
    x,y = p.get_xy()
    if height>0.01:
        ax.text(x+width/2,y+height/2,f'{height:.1%}',ha='center',va='center')
plt.show()
```

The analysis of payment methods highlights a dramatic variance in customer loyalty, with one method standing out as a major risk factor. Customers who pay via Electronic check have an exceptionally high churn rate of 45.3%, which is substantially higher than all other options. Conversely, customers enrolled in automatic payment methods, such as Credit card (15.2%) and Bank transfer (16.7%), demonstrate the highest retention rates. This indicates that automatic payment plans are a strong indicator of customer stability, while manual electronic payments signal a significant churn risk.

```python
fig,ax = plt.subplots(figsize=(5,6))
payment_type = prime_df.groupby('PaperlessBilling')['Churn'].value_counts(normalize=True).unstack()
payment_type.plot(kind='bar',stacked=True,ax=ax)
for p in ax.patches:
    width,height = p.get_width(),p.get_height()
    x,y = p.get_xy()
    if height>0.01:
        ax.text(x+width/2,y+height/2,f'{height:.1%}',ha='center',va='center')
plt.show()
```

This analysis reveals a strong correlation between the adoption of paperless billing and customer churn. A key finding is that customers who opt for paperless billing exhibit a significantly higher churn rate of 33.6%. This is more than double the 16.3% churn rate observed among customers who receive traditional paper bills. This suggests that the paperless billing segment may represent a less engaged or less loyal customer group. Consequently, these customers are a critical focus area for retention strategies aimed at mitigating this elevated churn risk.
